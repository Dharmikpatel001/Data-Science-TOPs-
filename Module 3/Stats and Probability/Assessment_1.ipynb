### What is the difference between Descriptive and Inferential Statistics?

- Descriptive Statistics just describe the data you already have. 
- They summarize facts, like the average score of a class or the number of people who liked a movie, using simple tools like charts and graphs.

###

- Inferential Statistics use a small piece of data (a sample) to make a smart guess or prediction about a much larger group (the population). 
- For example, doctors test a drug on 100 people to see if it will likely work for everyone.

- An investment analyst wants to assess the relationship between the stock prices of two 
companies to identify potential investment opportunities. Data: Let's consider the daily closing prices (in dollars) of Company A and Company B for a sample of 20 trading days: 
Company A: 45, 47, 48, 50, 52, 53, 55, 56, 58, 60, 62, 64, 65, 67, 69, 70, 72, 74, 76, 77 Company B: 52, 54, 55, 57, 59, 60, 61, 62, 64, 66, 67, 69, 71, 73, 74, 76, 78, 80, 82, 83
- Calculate the covariance and correlation between the stock prices of Company A and Company B. Interpret the value of the covariance and explain the nature of the relationship between the two stocks.


```python
company_a = [45, 47, 48, 50, 52, 53, 55, 56, 58, 60, 62, 64, 65, 67, 69, 70, 72, 74, 76, 77]
company_b = [52, 54, 55, 57, 59, 60, 61, 62, 64, 66, 67, 69, 71, 73, 74, 76, 78, 80, 82, 83]

covariance_matrix = np.cov(company_a, company_b)
covariance = covariance_matrix[0, 1]

correlation = np.corrcoef(company_a, company_b)[0, 1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```

- A survey was conducted to determine if ice cream flavor preference is related to gender. The following data was collected:



                Chocolate           Vanilla             Strawberry

        Male       20                15                   10   

        Female     25                20                   30

```python
from scipy.stats import chi2_contingency

observed = [[20, 15, 10],
            [25, 20, 30]]

chi2, p_value, dof, expected = chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("P-value:", p_value)
print("Expected frequencies:\n", expected)
```

- In a standardized test, the scores are normally distributed with a mean of 100 and a standard deviation of 15.
- What is the probability that a randomly selected student scores above 115?
- What percentage of students scored between 85 and 115?

```python
from scipy.stats import norm
mean = 100
std_dev = 15

# 1. Probability that score is above 115
prob_above_115 = 1 - norm.cdf(115, mean, std_dev)

# 2. Probability that score is between 85 and 115
prob_between = norm.cdf(115, mean, std_dev) - norm.cdf(85, mean, std_dev)

print("Probability score > 115:", prob_above_115)
print("Percentage between 85 and 115:", prob_between * 100)
```
